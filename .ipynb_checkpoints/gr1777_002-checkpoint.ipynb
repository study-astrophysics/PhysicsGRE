## 002. Classical Mechanics (Energy and Momentum)

### You should know about

- [Linear Momentum Conservation](https://en.wikipedia.org/wiki/Momentum#Conservation)

If the velocities of the particles are $u_1$ and $u_2$ before the interaction, and afterwards they are $v_1$ and $v_2$, then

$$ m_1u_1 + m_2u_2 = m_1v_1 + m_2v_2 $$

- [Work-energy Principle](https://en.wikipedia.org/wiki/Work_(physics)#Workâ€“energy_principle)

$$ W = \Delta E_k = \frac{1}{2}mv^2_2 - \frac{1}{2}mv^2_1 $$ 

### Solution

By the linear momentum conservation,

- x-direction

$$m_{1}v_{1,x} + m_{2}v_{2,x} = (m_{1} + m_{2})v_{x} $$

$$(0.5) \cdot 2 + (1.0)\cdot 0 = (0.5 + 1.5) v_{x} $$

$$v_{x} = \frac{2}{3}$$

- y-direction

$$m_{1}v_{1,y} + m_{2}v_{2,y} = (m_{1} + m_{2})v_{y}$$ 

$$ (0.5) \cdot 0 + (1.0) \cdot 1 = (0.5 + 1.5) v_{y}$$

$$v_{y} = \frac{2}{3}$$
 
Therefore, 

$$v = \sqrt{v^2_x + v^2_y} = \frac{2\sqrt{2}}{3}$$



Then, using the work-energy principle,

$$ W = \Delta K.E = \Delta \frac{1}{2}mv^2 $$

$$ W = \frac{1}{2}((0.5) \cdot 2^2 + 1 \cdot 1^2) - \frac{1}{2}\left((1.5) \cdot  \left(\frac{2\sqrt{2}}{3}\right)^2\right) = \frac{5}{6} J$$

If we assume that all energy generated by collision is converted to heat, then the heat energy produced by the collision is $\frac{5}{6} J$

### Answer

(E) $\frac{5}{6}J$
